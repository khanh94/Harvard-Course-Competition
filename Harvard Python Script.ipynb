```python
#Harvard script
import numpy as np
import pandas as pd
train_predictors = pd.read_csv('train_predictors.txt', sep=",", header=None)
train_labels = pd.read_csv('train_labels.txt', sep=",", header=None)
```

```python
train_predictors.rename(columns=lambda x: 'col' + str(x), inplace=True)
```

```python
train = pd.concat([train_predictors, train_labels], axis=1)
train['Label'] = train[0]
del train[0]
train
```

```python
test_predictors = pd.read_csv('test_predictors.txt', sep=',', header=None)
```

```python
test_predictors.rename(columns=lambda x: 'col' + str(x), inplace=True)
```

```python
test = test_predictors
test
```

```python
y_train = train['Label']
x_train = train.iloc[:,0:102]
x_train
```

```python
import xgboost as xgb
from sklearn.grid_search import GridSearchCV
```

```python
gbm = xgb.XGBClassifier(max_depth=3, n_estimators=5000, learning_rate=0.05).fit(x_train, y_train)
predictions = gbm.predict(test)
```
